# Flight Delay Data Preprocessing, Feature Engineering & Aggregation

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df_raw = pd.read_csv('dataset/Airline_Delay_Cause.csv')
print(f'Shape: {df_raw.shape}')
df_raw.head()
```

## Data Cleaning

```python
df = df_raw.copy()

# Remove duplicates
df = df.drop_duplicates()
print(f'Shape after removing duplicates: {df.shape}')
```

```python
# Rename columns for readability
df = df.rename(columns={
    'carrier_name': 'airline',
    'airport_name': 'airport_full_name',
    'arr_flights': 'total_flights',
    'arr_del15': 'delayed_flights',
    'arr_cancelled': 'cancelled_flights',
    'arr_diverted': 'diverted_flights',
    'carrier_ct': 'carrier_delay_count',
    'weather_ct': 'weather_delay_count',
    'nas_ct': 'nas_delay_count',
    'security_ct': 'security_delay_count',
    'late_aircraft_ct': 'late_aircraft_delay_count',
    'arr_delay': 'total_delay_minutes',
    'carrier_delay': 'carrier_delay_minutes',
    'weather_delay': 'weather_delay_minutes',
    'nas_delay': 'nas_delay_minutes',
    'security_delay': 'security_delay_minutes',
    'late_aircraft_delay': 'late_aircraft_delay_minutes'
})
print('Columns after renaming:')
print(df.columns.tolist())
```

```python
# Convert numeric columns and fill NaN with 0
numeric_cols = [
    'year', 'month', 'total_flights', 'delayed_flights',
    'cancelled_flights', 'diverted_flights',
    'carrier_delay_count', 'weather_delay_count', 'nas_delay_count',
    'security_delay_count', 'late_aircraft_delay_count',
    'total_delay_minutes', 'carrier_delay_minutes', 'weather_delay_minutes',
    'nas_delay_minutes', 'security_delay_minutes', 'late_aircraft_delay_minutes'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Remove rows where total_flights <= 0
df = df[df['total_flights'] > 0]
print(f'Shape after cleaning: {df.shape}')
df.dtypes
```

## Feature Engineering

```python
# Rate features
df['delay_rate'] = df['delayed_flights'] / df['total_flights']
df['cancellation_rate'] = df['cancelled_flights'] / df['total_flights']
df['diversion_rate'] = df['diverted_flights'] / df['total_flights']

# Average delay per delayed flight (handle division by zero)
df['avg_delay_per_delayed_flight'] = df['total_delay_minutes'] / df['delayed_flights'].replace(0, np.nan)
df['avg_delay_per_delayed_flight'] = df['avg_delay_per_delayed_flight'].fillna(0)

print('Rate features created')
df[['delay_rate', 'cancellation_rate', 'diversion_rate', 'avg_delay_per_delayed_flight']].describe()
```

```python
# Season and quarter
df['season'] = df['month'].map({
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
})

df['quarter'] = df['month'].apply(lambda m: (m - 1) // 3 + 1)

# High delay flag
df['is_high_delay'] = (df['delay_rate'] > 0.20).astype(int)

print(f'Season distribution:\n{df["season"].value_counts()}')
print(f'\nHigh delay flights: {df["is_high_delay"].sum()} / {len(df)}')
```

```python
# Delay cause percentages
delay_cause_cols = [
    'carrier_delay_minutes', 'weather_delay_minutes',
    'nas_delay_minutes', 'security_delay_minutes', 'late_aircraft_delay_minutes'
]

total_delay = df[delay_cause_cols].sum(axis=1).replace(0, np.nan)

for col in delay_cause_cols:
    pct_col = col.replace('_minutes', '_pct')
    df[pct_col] = (df[col] / total_delay * 100).fillna(0)

# Dominant delay cause
cause_names = ['carrier', 'weather', 'nas', 'security', 'late_aircraft']
df['dominant_delay_cause'] = df[delay_cause_cols].idxmax(axis=1).str.replace('_delay_minutes', '')
# Set to 'none' where no delays exist
df.loc[total_delay.isna(), 'dominant_delay_cause'] = 'none'

print(f'Dominant delay cause distribution:\n{df["dominant_delay_cause"].value_counts()}')
```

```python
print(f'Final cleaned dataset shape: {df.shape}')
print(f'Columns: {df.columns.tolist()}')
df.head()
```

## Data Aggregation

```python
# Airport-level aggregation
airport_agg = df.groupby(['airport', 'airport_full_name'], as_index=False).agg({
    'total_flights': 'sum',
    'delayed_flights': 'sum',
    'cancelled_flights': 'sum',
    'total_delay_minutes': 'sum'
})
airport_agg['delay_rate'] = airport_agg['delayed_flights'] / airport_agg['total_flights']
airport_agg['cancellation_rate'] = airport_agg['cancelled_flights'] / airport_agg['total_flights']
airport_agg['avg_delay_minutes'] = airport_agg['total_delay_minutes'] / airport_agg['delayed_flights'].replace(0, 1)

print(f'Airport aggregation: {airport_agg.shape}')
airport_agg.sort_values('delay_rate', ascending=False).head(10)
```

```python
# Airline-level aggregation
airline_agg = df.groupby(['carrier', 'airline'], as_index=False).agg({
    'total_flights': 'sum',
    'delayed_flights': 'sum',
    'cancelled_flights': 'sum',
    'total_delay_minutes': 'sum'
})
airline_agg['delay_rate'] = airline_agg['delayed_flights'] / airline_agg['total_flights']
airline_agg['cancellation_rate'] = airline_agg['cancelled_flights'] / airline_agg['total_flights']
airline_agg['avg_delay_minutes'] = airline_agg['total_delay_minutes'] / airline_agg['delayed_flights'].replace(0, 1)

print(f'Airline aggregation: {airline_agg.shape}')
airline_agg.sort_values('delay_rate', ascending=False).head(10)
```

```python
# Monthly trend aggregation
monthly_agg = df.groupby(['year', 'month'], as_index=False).agg({
    'total_flights': 'sum',
    'delayed_flights': 'sum',
    'delay_rate': 'mean'
})
monthly_agg = monthly_agg.sort_values(['year', 'month'])
monthly_agg['year_month'] = monthly_agg['year'].astype(int).astype(str) + '-' + monthly_agg['month'].astype(int).astype(str).str.zfill(2)

print(f'Monthly trends: {monthly_agg.shape}')
monthly_agg.head(12)
```

```python
# Seasonal aggregation
season_agg = df.groupby('season', as_index=False).agg({
    'total_flights': 'sum',
    'delayed_flights': 'sum',
    'delay_rate': 'mean'
})
season_order = ['Winter', 'Spring', 'Summer', 'Fall']
season_agg['season'] = pd.Categorical(season_agg['season'], categories=season_order, ordered=True)
season_agg = season_agg.sort_values('season')

print('Seasonal aggregation:')
season_agg
```

```python
# Airport-Airline cross aggregation
airport_airline_agg = df.groupby(['airport', 'carrier'], as_index=False).agg({
    'total_flights': 'sum',
    'delayed_flights': 'sum',
    'total_delay_minutes': 'sum'
})
airport_airline_agg['delay_rate'] = airport_airline_agg['delayed_flights'] / airport_airline_agg['total_flights']

print(f'Airport-Airline cross aggregation: {airport_airline_agg.shape}')
airport_airline_agg.sort_values('delay_rate', ascending=False).head(10)
```

## Save Processed Data

```python
df.to_csv('dataset/cleaned_flight_data.csv', index=False)
airport_agg.to_csv('dataset/airport_metrics.csv', index=False)
airline_agg.to_csv('dataset/airline_metrics.csv', index=False)
monthly_agg.to_csv('dataset/monthly_trends.csv', index=False)

print('All files saved to dataset/ folder:')
print('  - cleaned_flight_data.csv')
print('  - airport_metrics.csv')
print('  - airline_metrics.csv')
print('  - monthly_trends.csv')
```
